```python
import joblib
import pandas as pd

bundle = joblib.load("diabetes_model_with_threshold.pkl")

pipeline = bundle["pipeline"]
threshold = bundle["threshold"]

# Example new data
X_new = pd.DataFrame([{
    "Pregnancies": 6,
    "Glucose": 168,
    "BloodPressure": 72,
    "SkinThickness": 35,
    "Insulin": 130,
    "BMI": 33.6,
    "DiabetesPedigreeFunction": 0.63,
    "Age": 45
}])

prob = pipeline.predict_proba(X_new)[:, 1]
prediction = (prob >= threshold).astype(int)

print("Probability:", prob)
print("Prediction:", prediction)
```
